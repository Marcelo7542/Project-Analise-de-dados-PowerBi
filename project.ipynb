```python

import pandas as pd
dados = pd.read_csv('orders.csv', na_values=['Not Available','unknown'])

dados.columns = dados.columns.str.lower()
dados.rename(columns={
    'order id': 'ID do Pedido',
    'order date': 'Data do Pedido',
    'ship mode': 'Modo de Envio',
    'segment': 'Segmento',
    'country': 'País',
    'city': 'Cidade',
    'state': 'Estado',
    'postal code': 'Código Postal',
    'region': 'Região',
    'category': 'Categoria',
    'sub category': 'Subcategoria',
    'product id': 'ID do Produto',
    'cost price': 'Preço de Custo',
    'list price': 'Preço de Lista',
    'quantity': 'Quantidade',
    'discount percent': 'Percentual de Desconto',
}, inplace=True)
dados.columns = dados.columns.str.replace(' ','_')
# Substituir valores na coluna 'Modo_de_Envio'
dados['Modo_de_Envio'] = dados['Modo_de_Envio'].replace({
    'Second Class': 'Segunda Classe',
    'Standard Class': 'Classe Padrão',
    'First Class': 'Primeira Classe',
    'Same Day': 'Mesmo Dia'
})

# Substituir valores na coluna 'Segmento'
dados['Segmento'] = dados['Segmento'].replace({
    'Consumer': 'Consumidor',
    'Corporate': 'Corporativo',
    'Home Office': 'Home Office'
})

# Substituir valores na coluna 'País'
dados['País'] = dados['País'].replace({
    'United States': 'Estados Unidos'
})

# Substituir valores na coluna 'Região'
dados['Região'] = dados['Região'].replace({
    'South': 'Sul',
    'West': 'Oeste',
    'Central': 'Central',
    'East': 'Leste'
})

# Substituir valores na coluna 'Categoria'
dados['Categoria'] = dados['Categoria'].replace({
    'Furniture': 'Móveis',
    'Office Supplies': 'Material de Escritório',
    'Technology': 'Tecnologia'
})

# Substituir valores na coluna 'Subcategoria'
dados['Subcategoria'] = dados['Subcategoria'].replace({
    'Bookcases': 'Estantes',
    'Chairs': 'Cadeiras',
    'Labels': 'Etiquetas',
    'Tables': 'Mesas',
    'Storage': 'Armazenamento',
    'Furnishings': 'Mobiliário',
    'Art': 'Arte',
    'Phones': 'Telefones',
    'Binders': 'Pastas',
    'Appliances': 'Eletrodomésticos',
    'Paper': 'Papel',
    'Accessories': 'Acessórios',
    'Envelopes': 'Envelopes',
    'Fasteners': 'Fixadores',
    'Supplies': 'Suprimentos',
    'Machines': 'Máquinas',
    'Copiers': 'Copiadoras'
})
dados
```

```python
dados['Desconto']= dados['Preço_de_Lista']*dados['Percentual_de_Desconto']*.01
dados
```

```python
dados['Preço_da_venda'] = dados['Preço_de_Lista'] - dados['Desconto']
dados['Lucro'] = dados['Preço_da_venda'] - dados['Preço_de_Custo']
dados
```

```python
dados['Data_do_Pedido'] = pd.to_datetime(dados['Data_do_Pedido'],format="%Y-%m-%d")
dados
```

```python
dados.columns
dados.drop(columns=['Preço_de_Custo','Preço_de_Lista','Percentual_de_Desconto'], inplace=True)
dados
import unidecode
dados.columns = [unidecode.unidecode(col) for col in dados.columns]
print(dados.columns)
  # Exemplo com string

print(dados.dtypes)
```
